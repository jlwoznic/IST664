```python
nltk.corpus.gutenberg.fileids()
```

```python
myfile=nltk.corpus.gutenberg.fileids()[1]
```

```python
myfile
```

```python
persuasion=nltk.corpus.gutenberg.raw(myfile)
```

```python
persuaion_split=nltk.sent_tokenize(persuasion)
```

```python
persuasion_split_tokens=[nltk.word_tokenize(sent) for sent in persuaion_split]
```

```python
mystanford=[nltk.pos_tag(tokens) for tokens in persuasion_split_tokens]
```

```python
mystanford[:20]
```

```python
mysublist=[item for sublist in mystanford for item in sublist]
```

```python
tag_fd = nltk.FreqDist(tag for (word, tag) in mysublist)
```

```python
for tag,freq in tag_fd.most_common():
    print (tag, freq)
```
