# First Sample Deep Learning Model

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#Create Datapoints
from sklearn.datasets.samples_generator import make_circles
```

```python
X, y = make_circles(n_samples=1000,
                    noise=0.1,
                    factor=0.2,
                    random_state=0)
```

```python
X
```

```python
X.shape
```

```python
plt.figure(figsize=(5, 5))
plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.legend(['0', '1'])
plt.title("Blue circles and Red crosses")
```

```python
# Need to train a model to seperate the 2 classes


from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
```

```python
model = Sequential()
```

```python
model.add(Dense(4, input_shape=(2,), activation='tanh'))
```

```python
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(SGD(lr=0.5), 'binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=20)
```

```python
hticks = np.linspace(-1.5, 1.5, 101)
vticks = np.linspace(-1.5, 1.5, 101)
aa, bb = np.meshgrid(hticks, vticks)
ab = np.c_[aa.ravel(), bb.ravel()]
c = model.predict(ab)
cc = c.reshape(aa.shape)
```

```python
plt.figure(figsize=(5, 5))
plt.contourf(aa, bb, cc, cmap='bwr', alpha=0.2)
plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.legend(['0', '1'])
plt.title("Blue circles and Red crosses")
```

```python
def pretty_confusion_matrix(y_true, y_pred, labels=["False", "True"]):
    cm = confusion_matrix(y_true, y_pred) # inserte confustion Matrix into pandas dataframe
    pred_labels = ['Predicted '+ l for l in labels]
    df = pd.DataFrame(cm, index=labels, columns=pred_labels)
    return df
```

** Confusion Matrix as a Model Performance Measure **

Accuracy : (true Positive / True Negatives) / total

Precision : True Positive / Test Positives

Recall : True Positives / Actual Positives

FI Score = 2*(precision*Recall)/SUM(precision+ Recall)

FI will be close to 1 when precision and recal are close to 1, FI is good way to make sure both precision and recall are good


```python
# check the confusion matrix, precision and recall

pretty_confusion_matrix(y_test, y_test_pred, labels=['Stay', 'Leave'])
#1821 True negatives, 466 True Positives, 472 false postives, and 220 false negatives
#
```
