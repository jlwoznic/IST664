Each of 5 schools ( A, B, C, D, E )is implementing the same math course this semeseter, with 35 lessions. There are 30 sections total. The semester is about 3/4 of the way through.

For each section, we record the number of students who are : 

- Very Ahead ( more than 5 lessions ahead )
- Middling ( 5 - 0 ahead)
- behind (1 - 5 behind)
- more behind (6 - 10)
- very behind ( 10 + behind )
- Completed ( finished with course)

**What is the story with this data?**
